```python
import os
import re 
import sys
import urllib2
from bs4 import BeautifulSoup
```

```python
url_number = 735
url_base = "http://www.beeradvocate.com/beer/profile/"

url_name = url_base + str(url_number)

print url_name
```

```python
# all beer advocate breweries contain a numeric brewery ID which is not known
print "scraping: ",url_name
headers = { 'User-Agent' : 'Mozilla/5.0'  }
req = urllib2.Request(url_name, None, headers)
html = urllib2.urlopen(req).read()
soup = BeautifulSoup(html)
```

```python
print soup.prettify()
```

```python
for d in soup.findAll("table")[-1].findAll('tr')[-2].findAll("td"):
  print d.text
```

```python
table_rows = len(soup.findAll("table")[-1].findAll('tr'))
beer_attribs = []
stop_counter = 0
while stop_counter < table_rows:
    beer = []
    for attrib in soup.findAll("table")[-1].findAll('tr')[stop_counter].findAll("td"):
        beer.append(attrib.text)
    beer_attribs.append(beer)
    stop_counter += 1
    
for beers in beer_attribs:
    print " ".join(beers)
        
```
